```python
# @title load ps1
ps1 = pd.read_csv(r'https://raw.githubusercontent.com/RealThanosP/pred-main-mod/refs/heads/main/data/condition%2Bmonitoring%2Bof%2Bhydraulic%2Bsystems/PS1.txt', sep='\t', header = None)
ps1.head()
```

```python
ps1.shape
```

```python
ps1.describe()
```

```python
# @title pressure over time
import matplotlib.pyplot as plt

plt.plot(ps1.iloc[0])  # Πρώτη σειρά
plt.title("Pressure over time for first measurement")
plt.xlabel("Time (10 ms intervals)")
plt.ylabel("Pressure (bar)")
plt.show()
```

```python
# @title feature engineering
ps1_stats = pd.DataFrame({
    'mean': ps1.mean(axis=1),
    'std': ps1.std(axis=1),
    'min': ps1.min(axis=1),
    'max': ps1.max(axis=1),
    'range': ps1.max(axis=1) - ps1.min(axis=1),
    'rms': (ps1.pow(2).mean(axis=1)).pow(0.5)
})

ps1_stats
```

```python
# @title scalling
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
ps1_scaled = scaler.fit_transform(ps1)
```

```python
ps1_scaled
```

```python
ps1
```

```python
ps1_scaled = pd.DataFrame(ps1_scaled)
```

```python
ps1_scaled
```

```python
ps1_scaled_stats = pd.DataFrame({
    'mean': ps1_scaled.mean(axis=1),
    'std': ps1_scaled.std(axis=1),
    'min': ps1_scaled.min(axis=1),
    'max': ps1_scaled.max(axis=1),
    'range': ps1_scaled.max(axis=1) - ps1_scaled.min(axis=1),
    'rms': (ps1_scaled.pow(2).mean(axis=1)).pow(0.5)
})

ps1_scaled_stats
```

```python
ps2 = pd.read_csv(r'https://raw.githubusercontent.com/RealThanosP/pred-main-mod/refs/heads/main/data/condition%2Bmonitoring%2Bof%2Bhydraulic%2Bsystems/PS2.txt', sep='\t', header = None)
ps2.head()
```

```python
ps2_stats = pd.DataFrame({
    'mean': ps2.mean(axis=1),
    'std': ps2.std(axis=1),
    'min': ps2.min(axis=1),
    'max': ps2.max(axis=1),
    'range': ps2.max(axis=1) - ps2.min(axis=1),
    'rms': (ps2.pow(2).mean(axis=1)).pow(0.5)
})

ps2_stats
```

```python
# @title Combining Data
# it was a mistake to scale the data befome combing them, so i scaled them again afterwards
combined_stats1 = pd.concat([ps1_stats, ps2_stats], axis=1)
combined_stats1
```

```python
combined_stats1_scaled = scaler.fit_transform(combined_stats1)
combined_stats1_scaled = pd.DataFrame(combined_stats1_scaled)
combined_stats1_scaled
```

```python
combined_stats = pd.concat([combined_stats1_scaled, profile], axis=1)
combined_stats
#here i have to change the titles of the columns
```
